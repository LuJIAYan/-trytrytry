```python
filename="yes-minister.txt" #定义文件变量名
```

```python
mytext=open(filename,encoding='utf-8').read()  #读取文件
```

```python
mytext  #显示结果
```

```python
from wordcloud import WordCloud  #导入词云
```

```python
mycloud=WordCloud().generate(mytext)  #讲文本文件内容生成词云
```

```python
mycloud
```
